# Pre-processing School A

```python
A = pd.read_excel("平阳县怀溪镇晓坑中心学校数据.xlsx")
A
```

```python
A.describe()
```

```python
A.isnull().sum()
```

```python
A.drop(['家境','评价？'], axis=1, inplace = True)
```

```python
A.fillna(0, inplace = True)
A.head()
```

```python
A.rename(columns = {'是否资助':'Financial Grants', 
                       '年级':'Grade',
                      '性别': 'Gender',
                      '语文':'Chinese',
                      '数学':'Math',
                      '科学':'Science',
                      '英文':'English',
                      '社会':'Sociology',
                      '总分':'Total Score',
                      '收入':'Annual income',
                      '劳动力':'Work Force',
                      '家庭（父母）':'Parent status',
                      '日常生活':'Parental Care', 
                      '疾病/残疾':'Injuries/Sicknesses',
                      '欠款':'Debt',
                      '其他额外开销':'Other Expenses',
                      '学习态度':'Attitude towards study',
                      '被子':'Free Quilt',
                      '微心愿':'Charity',
                      '营养餐':'Free Lunch',
                      '生活费补助':'Daily Expenses Grants'}, inplace = True)

A
```

```python
A['Chinese'].replace(to_replace = 0, value = A['Chinese'].mean(), inplace=True)
A['Math'].replace(to_replace = 0, value = A['Math'].mean(), inplace=True)
A['Science'].replace(to_replace = 0, value = A['Science'].mean(), inplace=True)
A['English'].replace(to_replace = 0, value = A['English'].mean(), inplace=True)
A['Total Score'].replace(to_replace = 0, value = A['Total Score'].mean(), inplace=True)
A.head(10)
```

```python
A.replace({'Annual income' : {'中等' : 1, '低' : 0.66, '困难': 0.33, '低保': 0}}, inplace = True)
A.replace({'Parent status' : {'双' : 1, '父亲' : 0, '母亲' : 0}}, inplace = True)
A.replace({'Parental Care' : {'父母' : 1, '单亲' : 0.5, '离异' : 0.5, '留守' : 0}}, inplace = True)
A.replace({'Attitude towards study' : {'好' : 1, '一般' : 0.5, '差' : 0}}, inplace = True)
A.replace({'Gender' : {'m' : 1,  'f' : 0}}, inplace = True)
A
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
from sklearn.preprocessing import LabelEncoder
```

```python
A
```

```python
X = A.drop(['Math', 'Science', 'English', 'Chinese', 'Sociology', 'Total Score'], axis=1)
y_MathA = A['Math']
y_ScienceA = A['Science']
y_EnglishA = A['English']
y_ChineseA = A['Chinese']
y_SociologyA = A['Sociology']
y_TotalA = A['Total Score']

X_train, X_test, y_MathA_train, y_MathA_test = train_test_split(X, y_MathA, test_size=0.2, random_state=42)
X_train, X_test, y_ScienceA_train, y_ScienceA_test = train_test_split(X, y_ScienceA, test_size=0.2, random_state=42)
X_train, X_test, y_ScienceA_train, y_ScienceA_test = train_test_split(X, y_ScienceA, test_size=0.2, random_state=42)
X_train, X_test, y_EnglishA_train, y_EnglishA_test = train_test_split(X, y_EnglishA, test_size=0.2, random_state=42)
X_train, X_test, y_ChineseA_train, y_ChineseA_test = train_test_split(X, y_ChineseA, test_size=0.2, random_state=42)
X_train, X_test, y_SociologyA_train, y_SociologyA_test = train_test_split(X, y_SociologyA, test_size=0.2, random_state=42)
X_train, X_test, y_TotalA_train, y_TotalA_test = train_test_split(X, y_TotalA, test_size=0.2, random_state=42)
```

```python
X
```

## Normalization methods

### Z-score

```python
A_score = A[['Math','Chinese','English','Science','Total Score']]
  
# apply normalization techniques 
for column in A_score: 
    mean = np.mean(A_score[column])
    std_dev = np.std(A_score[column])
    z_scores = (A_score[column] - mean) / std_dev
    A_score[column] = z_scores

A_score
```

```python
B_score = B[['Math','Chinese','English','Science','Total Score']]
  
# apply normalization techniques 
for column in B_score: 
    mean = np.mean(B_score[column])
    std_dev = np.std(B_score[column])
    z_scores = (B_score[column] - mean) / std_dev
    B_score[column] = z_scores

B_score
```

### Decimal Scaling

```python
A_score = A[['Math','Chinese','English','Science','Total Score']]
  
# apply normalization techniques 
for column in A_score: 
    A_score[column] =A_score[column]  / A_score[column].abs().max() 
      
A_score
```

```python
B_score = B[['Math','Chinese','English','Science','Total Score']]
  
# apply normalization techniques 
for column in B_score: 
    B_score[column] = B_score[column]  / B_score[column].abs().max() 
      
B_score
```
